```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model('yolo_model.keras') 
```

```python
# Kamera akışını başlat
cap = cv2.VideoCapture(0)  

while True:
    # Kameradan bir kare yakala
    ret, frame = cap.read()
    if not ret:
        break

    # Görüntüyü modele uygun hale getirin
    # (örneğin, modeliniz 224x224 boyutunda girdi alıyorsa)
    input_image = cv2.resize(frame, (224, 224))  # Modelin girdi boyutuna göre ayarlayın
    input_image = input_image / 255.0  # Normalizasyon
    input_image = np.expand_dims(input_image, axis=0)  # (1, 224, 224, 3) boyutunda olmalı

    # Model ile tahmin yap
    predictions = model.predict(input_image)
    predicted_class = np.argmax(predictions, axis=1)[0]

    # Sonucu görüntü üzerinde göster
    cv2.putText(frame, f"Prediction: {predicted_class}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Tahmin edilen kareyi göster
    cv2.imshow('Real-time Object Detection', frame)

    # 'q' tuşuna basarak çık
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Kaynakları serbest bırak
cap.release()
cv2.destroyAllWindows()
```
